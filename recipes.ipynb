## Regression model

```python
model = LinearRegression(
    copy_X=True,
)

param = {
    "fit_intercept": [False, True],
    "positive": [False, True],
}

gs = GridSearchCV(
    model,
    param_grid=param,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    cv=5
)

gs.fit(X_train, y_train)

model1 = gs.best_estimator_

y_pred = model1.predict(X_test)

sqrt(mean_squared_error(y_test ,y_pred))
```

```python
def get_api_key ():
    load_dotenv()
    api_key = os.getenv('API_KEY')
    if not api_key:
        raise ValueError('No api key found')
    print(api_key)
    return api_key

def request_food(ingridients):
    url = f"https://api.nal.usda.gov/fdc/v1/foods/search"
    table = []
    index = 0
    for ingridient in ingridients:
        index += 1
        print (f"{index} from {len(ingridients)}")
        params = {
        'api_key' :get_api_key(),
        'query' : ingridient,
        'pageSize' : 1}
        time.sleep(0.5)
        try:
            response = requests.get(url, params=params, timeout=10)
            if response.status_code != 200:
                print(f'Error {response.status_code}')
                continue
        except:
            print ('error')
        if response.status_code == 200:
            data = response.json()
            row = {"name" : ingridient}
            if data.get('foods') and len(data['foods']) > 0:
                for i in data['foods'][0]['foodNutrients']:
                    row[i['nutrientName']] = i['value']
                table.append(row)
            else:
                table.append({'name': ingridient})
    if table:
        all_columns = set()
        for row in table:
            all_columns.update(row.keys())
            other_columns = sorted([col for col in all_columns if col != 'name'])
            all_columns_ordered = ['name'] + other_columns
        
        with open('foodNutrients.csv', 'w', encoding='utf-8', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=all_columns_ordered, quoting=csv.QUOTE_NONNUMERIC)
            writer.writeheader()
            writer.writerows(table)
    
    return table
```

```python

def process_nutrition_data(df):
    result = pd.DataFrame()
    simple_mapping = {
        'Protein': 'protein',
        'Carbohydrate, by difference': 'carbohydrat',
        'Total lipid (fat)': 'fat',
        'Iron, Fe': 'Fe',
        'Magnesium, Mg': 'Mg',
        'Sodium, Na': 'Na',
        'Potassium, K': 'K',
        'Calcium, Ca': 'Ca',
        'Zinc, Zn': 'Zn',
        'Selenium, Se': 'Se',
        'Chromium': 'Cr',
        'Iodine': 'I',
        'Vitamin C, total ascorbic acid': 'C',
        'Copper, Cu': 'Cu'
    }
    for old_name, new_name in simple_mapping.items():
        if old_name in df.columns:
            result[new_name] = df[old_name]
    d_columns = ['Vitamin D (D2 + D3)', 
                 'Vitamin D (D2 + D3), International Units',
                 'Vitamin D2 (ergocalciferol)',
                 'Vitamin D3 (cholecalciferol)']
    result['D'] = 0
    for col in d_columns:
        if col in df.columns:
            result['D'] = result['D'].fillna(0) + df[col].fillna(0)
        b12_columns = ['Vitamin B-12', 'Vitamin B-12, added']
    result['B12'] = 0
    for col in b12_columns:
        if col in df.columns:
            result['B12'] = result['B12'].fillna(0) + df[col].fillna(0)
    e_columns = ['Vitamin E (alpha-tocopherol)', 'Vitamin E, added']
    result['E'] = 0
    for col in e_columns:
        if col in df.columns:
            result['E'] = result['E'].fillna(0) + df[col].fillna(0)
    result['A'] = 0
    
    if 'Vitamin A, RAE' in df.columns:
        result['A'] = df['Vitamin A, RAE'].fillna(0)
    else:
        if 'Retinol' in df.columns:
            result['A'] = result['A'] + df['Retinol'].fillna(0)
        if 'Carotene, beta' in df.columns:
            result['A'] = result['A'] + df['Carotene, beta'].fillna(0) / 12
        if 'Carotene, alpha' in df.columns:
            result['A'] = result['A'] + df['Carotene, alpha'].fillna(0) / 24
        if 'Cryptoxanthin, beta' in df.columns:
            result['A'] = result['A'] + df['Cryptoxanthin, beta'].fillna(0) / 24
    k_columns = ['Vitamin K (phylloquinone)',
                 'Vitamin K (Dihydrophylloquinone)',
                 'Vitamin K (Menaquinone-4)']
    
    result['K'] = 0
    for col in k_columns:
        if col in df.columns:
            result['K'] = result['K'].fillna(0) + df[col].fillna(0)
    
    if 'name' in df.columns:
        result.insert(0, 'name', df['name'])
    
    return result
```
